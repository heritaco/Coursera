Sí. Guía exprés de **cálculo matricial** para ML/estadística.

# Objetos y formas

* Escalar $a\in\mathbb R$. Vector $x\in\mathbb R^{n}$. Matriz $A\in\mathbb R^{m\times n}$.
* Transpuesta $A^{\top}$. Simétrica $A=A^{\top}$. PSD $A\succeq0\iff x^{\top}Ax\ge0\ \forall x$.
* Traza $\operatorname{tr}(A)=\sum_i A_{ii}$. Determinante $\det A$. Inversa $A^{-1}$ si existe.

# Álgebra básica

* Producto: $Ax$ definido si $A:m\times n,\ x:n\to \mathbb R^{m}$.
* Traza: $\operatorname{tr}(AB)=\operatorname{tr}(BA)$ si los productos están definidos.
* Normas: $\|x\|_2=\sqrt{x^{\top}x}$, $\|A\|_F=\sqrt{\operatorname{tr}(A^{\top}A)}$.
* Proyección ortogonal: $P=X(X^{\top}X)^{-1}X^{\top}$ si $X$ tiene rango completo.

# Diferenciales (reglas mecánicas)

* $d(a)=0$. $d(x^{\top}y)=x^{\top}dy+y^{\top}dx$.
* $d(Ax)=A\,dx$ si $A$ constante.
* **Traza-truco**: $x^{\top}Ax=\operatorname{tr}(x^{\top}Ax)=\operatorname{tr}(Axx^{\top})$.
* Producto: $d(AB)=dA\,B+A\,dB$.
* Inversa: $d(A^{-1})=-A^{-1}(dA)A^{-1}$.
* Determinante: $d\log\det A=\operatorname{tr}(A^{-1}dA)$ para $A\succ0$.

# Gradientes más usados

Sea el gradiente respecto a $x$ columna y a $A$ de igual tamaño.

* $\nabla_x (a^{\top}x)=a$.
* $\nabla_x \tfrac12\|x\|_2^2=x$.
* $\nabla_x (x^{\top}Ax)= (A+A^{\top})x$. Si $A$ simétrica: $2Ax$.
* $\nabla_x \tfrac12\|Ax-b\|_2^2 = A^{\top}(Ax-b)$.
* $\nabla_A \tfrac12\|AX-B\|_F^2=(AX-B)X^{\top}$.
* $\nabla_X \operatorname{tr}(A^{\top}X)=A$.
* $\nabla_X \operatorname{tr}(X^{\top}AX)= (A+A^{\top})X$ (→ $2AX$ si $A$ simétrica).
* $\nabla_X \log\det X = (X^{-1})^{\top}$ para $X\succ0$.

# Hessianos típicos

* $f(x)=\tfrac12 x^{\top}Qx + c^{\top}x$: $\nabla f=Qx+c$, $\nabla^2 f=Q$.
* Convexidad: $f$ es convexa ⇔ $Q\succeq0$.

# Normal equations (LS)

$$
\min_w \tfrac12\|Xw-y\|^2 \Rightarrow \nabla_w= X^{\top}(Xw-y)=0
\Rightarrow w=(X^{\top}X)^{-1}X^{\top}y.
$$

# Operadores útiles

* **Vec**: $\mathrm{vec}(AXB)=(B^{\top}\otimes A)\mathrm{vec}(X)$.
* **Kronecker**: $(A\otimes B)(C\otimes D)=(AC)\otimes(BD)$.

# Cadenas y composición

Si $g(x)\in\mathbb R^{m}$, $f:\mathbb R^{m}\!\to\!\mathbb R$:
$\nabla_x f(g(x)) = J_g(x)^{\top}\nabla f$, con $J_g$ jacobiano de $g$.

# Identidades rápidas

* $\operatorname{tr}(ABC)=\operatorname{tr}(BCA)=\operatorname{tr}(CAB)$.
* $\|Ax\|_2^2 = x^{\top}A^{\top}Ax$.
* $d\|X\|_F^2 = 2\operatorname{tr}(X^{\top}dX)$.

# Mini-ejercicios (clave en una línea)

1. Deriva $f(w)=\tfrac12\|Aw-b\|^2$. → $A^{\top}(Aw-b)$.
2. Deriva $g(X)=\operatorname{tr}(X^{\top}AX)+b^{\top}\mathrm{vec}(X)$. → $(A+A^{\top})X+\mathrm{unvec}(b)$.
3. Deriva $h(X)= -\log\det X + \tfrac12\|CX-I\|_F^2$ con $X\succ0$. → $-(X^{-1})^{\top}+C^{\top}(CX-I)$.
