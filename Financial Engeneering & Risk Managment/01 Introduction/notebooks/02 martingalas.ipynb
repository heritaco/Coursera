### 1. El s√≠mbolo $P$

Ese super√≠ndice $E^P$ significa que la **esperanza se toma bajo la probabilidad $P$**.

### 2. Definici√≥n intuitiva

Un proceso $\{X_n\}$ es martingala si:

## 2. Varianza de $X_n$

* Para una sola variable:

  $$
  \mathrm{Var}(Z) = \mathbb{E}[Z^2] - (\mathbb{E}[Z])^2 = 1 - 0 = 1,
