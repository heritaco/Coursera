Sí. Idea central: **paridad put–call** es una identidad de *payoffs terminales*. Funciona en europeas porque **no hay ejercicio anticipado**. Falla como igualdad en americanas porque el holder puede **alterar los flujos antes del vencimiento**.

---

## 1) Paridad europea = igualdad exacta

Sin dividendos:

$$
C_E-P_E=S_0-\frac{K}{(1+r)^T}.
$$

Con dividendos discretos PV $(D)$ o rendimiento continuo $q$:

$$
C_E-P_E=S_0-\text{PV}(D)-\frac{K}{(1+r)^T}
\quad\text{o}\quad
C_E-P_E=S_0e^{-qT}-Ke^{-rT}.
$$

### Por qué vale

Replica el **mismo payoff en $T$** de dos maneras:

* Portafolio A: **call** + **bono** que pagará $K$ en $T$.
* Portafolio B: **put** + **acción** (menos dividendos, si aplica).

En $T$:
$\max(S_T-K,0)+K=\max(K-S_T,0)+S_T$ siempre.
Mismo payoff ⇒ **mismo precio hoy**. La clave: *nadie puede ejercer antes*.

---

## 2) Paridad americana = solo **desigualdades**

La posibilidad de **ejercicio temprano** rompe la igualdad de flujos *antes* de $T$. Resultado: **bandas**.

### Sin dividendos

* Límite inferior:

$$
C_A-P_A\ \ge\ (S_0-K).
$$

Prueba corta: $C_A\ge(S_0-K)^+$, $P_A\ge(K-S_0)^+$. Entonces
$(S_0-K)^+-(K-S_0)^+=S_0-K$.

* Límite superior:

$$
C_A-P_A\ \le\ S_0-\frac{K}{(1+r)^T}.
$$

Porque para acción sin dividendos **no conviene** ejercer el call temprano ⇒ $C_A=C_E$. Además $P_A\ge P_E$. Luego

$$
C_A-P_A\le C_E-P_E=S_0-\frac{K}{(1+r)^T}.
$$

En resumen, sin dividendos:

$$
\boxed{\,S_0-K\ \le\ C_A-P_A\ \le\ S_0-\tfrac{K}{(1+r)^T}\, }.
$$

Solo se vuelve **igualdad** si, además, no hay incentivo de ejercer el put temprano (casos especiales). En general, **no**.

### Con dividendos

Los dividendos hacen posible ejercer anticipadamente el **call** antes de la fecha ex–div. Las bandas se ajustan restando la PV de dividendos:

$$
\boxed{\,S_0-\text{PV}(D)-K\ \le\ C_A-P_A\ \le\ S_0-\text{PV}(D)-\tfrac{K}{(1+r)^T}\, }.
$$

Aquí ya no vale $C_A=C_E$; el call americano puede valer **más** que el europeo.

---

## 3) Micro–intuición

* **Europeas:** fijan el payoff en $T$. Dos carteras que igualan el payoff en $T$ deben valer lo mismo hoy.
* **Americanas:** el derecho de **timing** agrega un componente de valor que puede activarse en fechas distintas en call y put. Como sus cronogramas pueden divergir, solo podemos acotar el **spread** $C_A-P_A$, no fijarlo exactamente.

---

## 4) Mini–ejemplo numérico (sin dividendos)

$S_0=100,\ K=100,\ r=5\%,\ T=1$.

* Paridad europea: $C_E-P_E=100-95.24=4.76$.
* Bandas americanas: $100-100=0\ \le C_A-P_A\ \le 4.76$.

Si el put americano es muy valioso por posible ejercicio temprano, $P_A$ sube y el **spread** cae, pero nunca baja de $0$. Sin early exercise útil, el spread puede acercarse a $4.76$, pero **no** lo excede.
